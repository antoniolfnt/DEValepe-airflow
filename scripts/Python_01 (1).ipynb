```python
HEADER_ROW = 3
INITIAL_COLUMN_NAME = uuid.uuid4().hex
ROW_INDEX_COLUMN_NAME = uuid.uuid4().hex
SEPARATOR = ";"
```

```python
df = spark.read.text("/FileStore/tables/landing/Despesas_detalhadas_2008-1.csv")
```

```python
df = df\
  .withColumn(ROW_INDEX_COLUMN_NAME, F.monotonically_increasing_id())\
  .filter(F.col(ROW_INDEX_COLUMN_NAME) >= (HEADER_ROW - 1))\
  .drop(ROW_INDEX_COLUMN_NAME)\
  .withColumnRenamed(df.columns[0], INITIAL_COLUMN_NAME)
```

```python
header_columns = df.limit(1).collect()[0][INITIAL_COLUMN_NAME].split(";")
```

```python
for index, column in enumerate(header_columns):
  df = df.withColumn(
    column,
    F.split(F.col(INITIAL_COLUMN_NAME), SEPARATOR).getItem(index)
  )
```

```python
df = df.filter(~(F.col(df.columns[1]) == df.columns[1])).drop(INITIAL_COLUMN_NAME)
```

```python
display(df)
```

```python
final_df = df
display(final_df)
```
